```
plt.figure(figsize=(12, 7))
import matplotlib.patheffects as path_effects
def add_median_labels(ax):
    lines = ax.get_lines()
    # determine number of lines per box (this varies with/without fliers)
    boxes = [c for c in ax.get_children() if type(c).__name__ == 'PathPatch']
    lines_per_box = int(len(lines) / len(boxes))
    # iterate over median lines
    for median in lines[4:len(lines):lines_per_box]:
        # display median value at center of median line
        x, y = (data.mean() for data in median.get_data())
        # choose value depending on horizontal or vertical plot orientation
        value = x if (median.get_xdata()[1]-median.get_xdata()[0]) == 0 else y
        text = ax.text(x, y, f'{value:.1f}', ha='center', va='center',
                       fontweight='bold', color='white')
        # create median-colored border around white text for contrast
        text.set_path_effects([
            path_effects.Stroke(linewidth=3, foreground=median.get_color()),
            path_effects.Normal(),
        ])

box_plot=sns.boxplot(x='Gender',y='Age', hue='SepsisLabel',data=filter_train)
medians = filter_train.groupby(['SepsisLabel'])['Age'].median()
vertical_offset = filter_train['Age'].median() * 0.05 # offset from median for display
add_median_labels(box_plot.axes)
plt.show()
```

```
def impute_age(cols):
    Age = cols[0]
    SepsisLabel = cols[1]
    
    if pd.isnull(Age):

        if SepsisLabel == 0:
            return 65

        elif SepsisLabel  == 1:
            return 64

    else:
        return Age
```

**feature engineer HR**


```
def fe_hr(filter_train):
    filter_train.loc[(filter_train['HR'] >= 100) & (filter_train['Age'] >= 10 ),
            'custom_hr'] = 'abnormal'
    filter_train.loc[(filter_train['HR'] < 100) & (filter_train['HR'] > 60) & (filter_train['Age'] >= 10 ),
            'custom_hr'] = 'normal'
    filter_train.loc[(filter_train['HR'] >= 70) & (filter_train['HR'] < 190) & (filter_train['Age'] < 10 ),
            'custom_hr'] = 'normal'
    filter_train.loc[((filter_train['HR'] < 70) | (filter_train['HR'] >= 190)) & (filter_train['Age'] < 10 ),
            'custom_hr'] = 'abnormal'
    return filter_train
```

**Temp**

```
def fe_temp(filter_train):
    filter_train.loc[(filter_train['Temp'] >= 36.4) & (filter_train['Temp'] < 37.6), 
            'custom_temp'] = 'normal'
    filter_train.loc[(filter_train['Temp'] < 36.4) | (filter_train['Temp'] >= 37.6), 
            'custom_temp'] = 'abnormal'

    
    return filter_train
```

```
def fe_age(filter_train):
    filter_train.loc[filter_train['Age'] >=65, 'custom_age'] = 'old'
    filter_train.loc[filter_train['Age'] <1, 'custom_age'] = 'infant'
    filter_train.loc[(filter_train['Age'] >=1) & (filter_train['Age'] <65), 
            'custom_age'] = 'child/adult'
    return filter_train
```

```
def fe_o2sat(filter_train):
    filter_train.loc[(filter_train['O2Sat'] >= 90) & (filter_train['O2Sat'] < 100), 
            'custom_o2sat'] = 'normal'
    filter_train.loc[(filter_train['O2Sat'] < 90) & (filter_train['O2Sat'] >= 0), 
            'custom_o2sat'] = 'abnormal'


    return filter_train
```

```
def fe_bp(filter_train):
    filter_train.loc[(filter_train['SBP'] <90) & (filter_train['DBP'] <60), 'custom_bp'] = 'low'

    filter_train.loc[(filter_train['SBP'].between(90,120, inclusive=True)) & 
            (filter_train['DBP'].between(60,80, inclusive=True)), 
            'custom_bp'] = 'normal'


    filter_train.loc[(filter_train['SBP'].between(120,140, inclusive=True)) & 
            (filter_train['DBP'].between(80,90, inclusive=True)), 
            'custom_bp'] = 'elevated'


    filter_train.loc[(filter_train['SBP'] > 140 ) & 
            (filter_train['DBP'] > 90 ), 'custom_bp'] = 'high'

  
    return filter_train
```

```
def fe_rr(filter_train):
    filter_train.loc[(filter_train['Resp'].between(30,60)) & (filter_train['Age'] <1), 
            'custom_resp'] = 'normal'
    filter_train.loc[((filter_train['Resp'] < 30) | (filter_train['Resp'] > 60)) & 
             (filter_train['Age'] <1) ,'custom_resp'] = 'abnormal'


    filter_train.loc[(filter_train['Resp'].between(24,40)) & (filter_train['Age'].between(1,3)), 
            'custom_resp'] = 'normal'
    filter_train.loc[((filter_train['Resp'] < 24) | (filter_train['Resp'] > 40)) & 
             (filter_train['Age'].between(1,3)) ,'custom_resp'] = 'abnormal'


    filter_train.loc[(filter_train['Resp'].between(22,34)) & (filter_train['Age'].between(3,6)), 
            'custom_resp'] = 'normal'
    filter_train.loc[((filter_train['Resp'] < 22) | (filter_train['Resp'] > 34)) & 
             (filter_train['Age'].between(3,6)) ,'custom_resp'] = 'abnormal'


    filter_train.loc[(filter_train['Resp'].between(18,30)) & (filter_train['Age'].between(6,12)), 
            'custom_resp'] = 'normal'
    filter_train.loc[((filter_train['Resp'] < 18) | (filter_train['Resp'] > 30)) & 
             (filter_train['Age'].between(6,12)) ,'custom_resp'] = 'abnormal'


    filter_train.loc[(filter_train['Resp'].between(12,20)) & (filter_train['Age'] >12), 
            'custom_resp'] = 'normal'
    filter_train.loc[((filter_train['Resp'] < 12) | (filter_train['Resp'] > 20)) & (filter_train['Age'] >12),
            'custom_resp'] = 'abnormal'


    return filter_train
```
