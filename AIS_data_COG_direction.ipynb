```python
df1= pd.read_csv(path + 'ais_data_geohash_version3.csv')
```

```python
def categorize(direction):
  if direction['COG']<22.5:
    return 'N'
  if direction['COG']>=22.5 and direction['COG']<67.5:
    return 'NE'
  if direction['COG']>=67.5 and direction['COG']<112.5:
    return 'E'
  if direction['COG']>=112.5 and direction['COG']<157.5:
    return 'SE'
  if direction['COG']>=157.5 and direction['COG']<202.5:
    return 'S'
  if direction['COG']>=202.5 and direction['COG']<247.5:
    return 'SW'
  if direction['COG']>=247.5 and direction['COG']<292.5:
    return 'W'
  if direction['COG']>=292.5 and direction['COG']<337.5:
    return 'NW'
  if direction['COG']>=337.5:
    return 'N'
```

```python
df1['COG_direction'] = df1.apply(lambda direction: categorize(direction), axis=1)
```
