```python
# variable to hold the count of MMSI
count = 0
# list to hold MMSI unique values
mmsi_values =[]
#loop to count unique MMSI values
for i in range(0, len(df['MMSI'])):
  if df['MMSI'][i] not in mmsi_values:
    mmsi_values.append(df['MMSI'][i])
    count+=1
print('No of unique MMSI values: ', count)
print('MMSI unique values: ',mmsi_values)
```

```python
mmsi_values_csv = pd.DataFrame(data={"col1": mmsi_values})
mmsi_values_csv.to_csv(path+'mmsi_values.csv', sep=',', index=False)
```
